```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
x_data = np.r_[np.random.randn(20, 2) - [2,2], np.random.randn(20,2) + [2,2]]
y_data = [0] * 20 + [1] * 20

plt.scatter(x_data[:, 0], x_data[:, 1], c = y_data)
```

```python
model = svm.SVC(kernel='linear')
model.fit(x_data, y_data)
```

```python
print(model.n_support_)
```

```python
print(model.support_)
```

```python
print(model.support_vectors_)
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
b = model.intercept_
k1 = model.coef_[0][0]
k2 = model.coef_[0][1]

x_test = np.array([[-5],[5]])
y_test = - b / k2 - k1 * x_test / k2

plt.plot(x_test, y_test, 'k')
plt.scatter(x_data[:, 0], x_data[:,1], c = y_data)
```

```python
# plot the decision boundary and the support lines
point_1 = model.support_vectors_[0]
point_2 = model.support_vectors_[-1]

y_down = - k1 * x_test / k2 + k1 * point_1[0] / k2 + point_1[1] 
y_up = - k1 * x_test / k2 + k1 * point_2[0] / k2 + point_2[1] 

plt.scatter(x_data[:,0], x_data[:,1], c = y_data)
plt.plot(x_test, y_test, 'k')
plt.plot(x_test, y_down, 'r--')
plt.plot(x_test, y_up, 'b--')
```
