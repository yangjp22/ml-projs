```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
# set(): reset the default parameters
sns.set()
sinplot()
```

### 5种主题风格 ###
* darkgrid
* whitegrid
* dark
* white
* ticks
*** sns.set_style()

```python
sns.set_style('whitegrid')
data = np.random.normal(size = (20,6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

### 移除轴 despine ### 

```python
# remove the axes by despine(), offset shift the x,y axises
sinplot()
sns.despine(offset = 4)
```

```python
#f, ax = plt.subplots()
sns.violinplot(data)
sns.despine(offset=10)
```

```python
sns.set_style("whitegrid")
sns.boxplot(data = data,palette='deep')
sns.despine(left=True)
```

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

### 线条格式设置 ###
* paper
* poster
* talk
* notebook

```python
# reset the default parameters
sns.set()
sns.set_context("paper")
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context("talk")
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context("poster")
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
sinplot()
```
