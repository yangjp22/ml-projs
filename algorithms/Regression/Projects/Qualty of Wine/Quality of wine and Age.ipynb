```python
%matplotlib notebook
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# loading dataset
data = np.genfromtxt('./Wine.csv', delimiter=',')

# plot the data
plt.scatter(data[1:, 0], data[1:, 1])
plt.title('Quality Vs Age')
plt.xlabel('Age')
plt.ylabel('Quality')
```

```python
# split the dataset into training set and testing set
x_train, x_test, y_train, y_test = train_test_split(data[1:, 0], data[1:, 1], test_size = 0.3, random_state = 1234)
# adding a dimension for dataset 
x_train = x_train[:, np.newaxis]
x_test = x_test[:, np.newaxis]
```

```python
model = LinearRegression(normalize = True)
model.fit(x_train, y_train)
```

```python
# plot the training set
plt.scatter(x_train, y_train, color = 'b')
# plot the predictions of training set
plt.plot(x_train, model.predict(x_train), color = 'red')
# adding title and xy labels
plt.title('Quality Vs Age')
plt.xlabel('Age')
plt.ylabel('Quality')
```

```python
# plot the testing set
plt.scatter(x_test, y_test, color = 'b')
# plot the predictions of testing set
plt.plot(x_test,model.predict(x_test), color ='r', linewidth=5)
# adding title and xy labels
plt.title('Age Vs Quality (Test set)')
plt.xlabel('Age')
plt.ylabel('Quality')
plt.show()
```
