```python
%matplotlib notebook
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Features, the first row is the constant item
X = np.array([[1,3,3],
              [1,4,3],
              [1,1,1],
              [1,0,2]])
# labels
Y = np.array([[1],
              [1],
              [-1],
              [-1]])
plt.scatter(X[:,1], X[:,2], c = Y[:,0])
```

```python
def update(x_data, y_data, Ws, Lr):
    '''
    Instruction: to update the weights
    
    Parameters: 
        x_data: the input features
        y_data: the real label
        Ws: the weights that will be updated
        Lr: learning rate
    
    Return:
        new_Ws: the weights after updation
    '''
    output = np.sign(np.dot(x_data,Ws))
    Ws_C = Lr * (x_data.T.dot(y_data - output))/ int(x_data.shape[0])
    new_Ws = Ws + Ws_C
    return new_Ws
```

```python
# initialize the weights, from -1 to 1, 3 x 1
Ws = (np.random.random([3,1]) - 0.5) * 2
```

```python
# update the Weights
# set the learning rate is 0.11
lr = 0.11
for _ in range(1000):
    Ws = update(X, Y, Ws, lr)

# plot the decision boundary

# positive point
x1 = [3,4]
y1 = [3,3]
# negative point
x2 = [1,0]
y2 = [1,2]

# calculate the interception and slope
k = - Ws[1] / Ws[2]
d = - Ws[0] / Ws[2]
print('k=',k)
print('d=',d)

xdata = (0,5)

plt.figure()
plt.plot(xdata, xdata * k + d,'r')
plt.scatter(x1,y1,c='b')
plt.scatter(x2,y2,c='y')
```
